```python
import pandas as pd

# Read the first dataframe from CSV
df1 = pd.read_csv("source/train_data/sample/2022.csv")

# Read the second dataframe from CSV
df2 = pd.read_csv("source/test_data/sample/2022.csv")

# Concatenate the dataframes vertically
concatenated_df_X = pd.concat([df1, df2], axis=0)

# Display the concatenated dataframe
concatenated_df_X
```

```python
y1 = pd.read_csv("source/train_data/label/label_2022.csv")
y2 = pd.read_csv("source/test_data/label/test_label.csv")

# Rename the 'label' column in y2 to 'label_2022'
y2 = y2.rename(columns={'label': 'label_2022'})

concatenated_df_y = pd.concat([y1, y2], axis=0)
concatenated_df_y
```

```python
concatenated_df = pd.concat([concatenated_df_X, concatenated_df_y], axis=1)
concatenated_df
```

```python
from sklearn.model_selection import train_test_split

# Perform stratified train-test split
train_df, test_df = train_test_split(concatenated_df, test_size=0.20, stratify=concatenated_df['label_2022'])

# Get the CIK for training set
train_cik = train_df['cik'].tolist()

# Get the CIK for test set
test_cik = test_df['cik'].tolist()

# Print the CIK for training set
print("CIK for training set:")
print(train_cik)

# Print the CIK for test set
print("CIK for test set:")
print(test_cik)
```

```python
for year in range(2017, 2023):
    df1 = pd.read_csv(f"source/train_data/sample/{year}.csv")
    df2 = pd.read_csv(f"source/test_data/sample/{year}.csv")
    df = pd.concat([df1, df2], axis=0)
    train_df = df[df["cik"].isin(train_cik)]
    train_df = train_df.sort_values('cik')
    test_df = df[df["cik"].isin(test_cik)]
    test_df = test_df.sort_values('cik')
    train_df.to_csv(f"source/train_data_2/sample/{year}.csv", index=False)
    test_df.to_csv(f"source/test_data_2/sample/{year}.csv", index=False)
```
