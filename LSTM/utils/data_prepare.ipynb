```python
import pandas as pd
import numpy as np
```

```python
y_path = "/Users/weichentao/Documents/USC/2023fall/540/project/topic/train_change_rate.csv"
y_df = pd.read_csv(y_path)
y_df
```

```python
with open("source/intersection_of_cik_in_training_set.txt") as f:
    cik_in_training_set = [int(line.strip()) for line in f.readlines()]
    cik_in_training_set = set(cik_in_training_set)

y_df = y_df[y_df["cik"].isin(cik_in_training_set)]
y_df
```

```python
years = range(2017, 2023)
for year in years:
    column_name = f"label_{year}"
    y_df[column_name] = y_df[f"{year}"].apply(lambda x: 1 if x > 0 else 0)
y_df
```

```python
y_df = y_df.sort_values('cik')
y_df
```

```python
column_names = y_df.columns[-6:].tolist()
```

```python
folder_path = "source/label"

for column_name in column_names:
    file_path = folder_path + "/" + column_name + ".csv"
    y_df[[column_name]].to_csv(file_path, index=False)
```
